```python
import pandas as pd
import numpy as np

# Ш§Щ„Ш±Ш§ШЁШ· Ш§Щ„Щ…ШЁШ§ШґШ± Щ„Щ„ШЇШ§ШЄШ§
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"

column_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigree', 'Age', 'Outcome']

# Щ‚Ш±Ш§ШЎШ© Ш§Щ„ШЇШ§ШЄШ§
df = pd.read_csv(url, names=column_names)

data = df.values

input_data = data[:, 0:8]
output_data = data[:, 8]

```

```python
### 2. Build Machine Learning Model

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
# ШЁЩ†Ш§ШЎ Щ‡ЩЉЩѓЩ„ Ш§Щ„ШґШЁЩѓШ© Ш§Щ„Ш№ШµШЁЩЉШ©
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu')) # Ш§Щ„Ш·ШЁЩ‚Ш© Ш§Щ„ШЈЩ€Щ„Щ‰ (8 Щ…ШЇШ®Щ„Ш§ШЄ)
model.add(Dense(8, activation='relu'))              # Ш§Щ„Ш·ШЁЩ‚Ш© Ш§Щ„Ш«Ш§Щ†ЩЉШ©
model.add(Dense(1, activation='sigmoid'))           # Ш·ШЁЩ‚Ш© Ш§Щ„Щ…Ш®Ш±Ш¬Ш§ШЄ (ШЄШµЩ†ЩЉЩЃ 0 ШЈЩ€ 1)
```

```python
###3. Set The Configuration
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

```

```python
### 4. Train The ML Model

model.fit(input_data, output_data, epochs=150, batch_size=10 )
```

```python
# Ш­ШіШ§ШЁ Ш§Щ„ШЇЩ‚Ш© Щ€ШЄЩ‚ЩЉЩЉЩ… Ш§Щ„Щ…Щ€ШЇЩЉЩ„
accuracy = model.evaluate(input_data, output_data, verbose=0)

# Ш·ШЁШ§Ш№Ш© Ш§Щ„Щ†ШЄЩЉШ¬Ш© (accuracy[1] ШЁШЄШ¬ЩЉШЁ Щ†ШіШЁШ© Ш§Щ„ШЇЩ‚Ш© Щ…Щ† Ш§Щ„Щ…ШµЩЃЩ€ЩЃШ©)
print("Accuracy: %.2f %%" % (accuracy[1] * 100))
```

```python
###2. (Make Predictions)
predictions = model.predict(input_data)

print(predictions[:5])
```

```python
def display(num):
    if num == 0:
        return "No Disease"
    else:
        return "Disease"

# Ш№Ш±Ш¶ ШЈЩ€Щ„ 10 Щ†ШЄШ§Ш¦Ш¬ Щ„Щ„Щ…Щ‚Ш§Ш±Щ†Ш© ШЁЩЉЩ† Ш§Щ„ШЄЩ€Щ‚Ш№ Щ€Ш§Щ„Ш­Щ‚ЩЉЩ‚Ш©
for i in range(10):
    print(" %s => %s (expected %s) " % (input_data[i].tolist(), display(predictions[i]), display(output_data[i])))
```
